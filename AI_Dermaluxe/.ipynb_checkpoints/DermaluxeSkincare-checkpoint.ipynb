```python
from sklearn.model_selection import GridSearchCV

max_depth = [None, 2,4,6,8,10,12]
parameters = {"max_depth": max_depth}

regressor = DecisionTreeRegressor(random_state=0)
gs = GridSearchCV(regressor, parameters, scoring='neg_mean_squared_error')
gs.fit(x, y.values)
```

```python
regressor = gs.best_estimator_

regressor.fit(x, y.values)
y_pred = regressor.predict(x)
error = np.sqrt(mean_squared_error(y, y_pred))
```

```python
from sklearn.preprocessing import LabelEncoder

# Example data used to train the LabelEncoders
age = ['16-19', '20-25', '26-30', '31 Above']
gender = ['Female', 'Male']
skin_type = ['Dry', 'Normal', 'Oily', 'Combination', 'Sensitive']
skin_tone = ['Fair', 'Medium', 'Deep', 'Dark', 'Olive']
skin_concern = ['Acne, Enlarged Pores', 'Redness, Dehydration','Hyperpigmentation', 'Aging, Fine Lines, Dullness', 'Uneven Skin Tone', 'Wrinkles, Sagging Skin']
environmental_impact = ['Spring – Moderate climate', 'Summer – Hot and humid','Autumn (Fall) – High pollution levels']
skin_goals = ['Reduce acne and blemishes', 'Improve hydration','Even out skin tone','Prevent aging signs','General maintenance']


# Create LabelEncoders
le_age = LabelEncoder()
le_age.fit(age)

le_gender = LabelEncoder()
le_gender.fit(gender)

le_skin_type = LabelEncoder()
le_skin_type.fit(skin_type)

le_skin_tone = LabelEncoder()
le_skin_tone.fit(skin_tone)

le_skin_concern = LabelEncoder()
le_skin_concern.fit(skin_concern)

le_environmental_impact = LabelEncoder()
le_environmental_impact.fit(environmental_impact)

le_skin_goals = LabelEncoder()
le_skin_goals.fit(skin_goals)
```

```python
import numpy as np

x = np.array([["16-19","Female","Dry", "Fair", "Acne, Enlarged Pores","Spring – Moderate climate","Reduce acne and blemishes"]])

# Apply the transformations
x[:, 0] = le_age.transform(x[:, 0]) 
x[:, 1] = le_gender.transform(x[:, 1])# Age
x[:, 2] = le_skin_type.transform(x[:, 2]) # Blood sugar level
x[:, 3] = le_skin_tone.transform(x[:, 3])   # Drugs
x[:, 4] = le_skin_concern.transform(x[:, 4])
x[:, 5] = le_environmental_impact.transform(x[:, 5])
x[:, 6] = le_skin_goals.transform(x[:, 6])
# Sitting time

print(x)
```
